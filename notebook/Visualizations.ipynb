# Activation Functions

## RELU

```python
import numpy as np

xs = np.arange(-1,1,0.1)
ys = np.maximum(xs,0)

import matplotlib.pyplot as plt
plt.plot(xs, ys)
plt.show()
```

## Tanh

```python
import numpy as np

xs = np.arange(-4,4,0.1)
ys = np.tanh(xs)

import matplotlib.pyplot as plt
plt.plot(xs, ys)
plt.show()
```
